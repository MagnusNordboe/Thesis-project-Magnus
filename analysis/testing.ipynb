```python
#Find the length of each individual time series
print("still works here")
import pandas as pd
arrays = [['A', 'A', 'B', 'B', 'C', 'C'],
          [1, 2, 1, 2, 1, 2]]
tuples = list(zip(*arrays))
index = pd.MultiIndex.from_tuples(tuples, names=['first', 'second'])
data = {'value': [1, 2, 3, 4, 5, 6]}
df = pd.DataFrame(data, index=index)
print("before func")
def padding(df:pd.DataFrame):
    print("started funct")
    instances = df.index.get_level_values(0).unique()
    print("found instances")
    for instance in instances:
        print("instance for loop")
        #This line of code gets every second level index value that corresponds to vthe current first level index value. 
        second_level_values = df.index.get_level_values(1)[df.index.get_level_values(0)==instance]
        print("got the values together")
        length = len(second_level_values)
        print("got the length")
        print("{} length of values is: {}".format(instance, length))


padding(imputed_df)
```
