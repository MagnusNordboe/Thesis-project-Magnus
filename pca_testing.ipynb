```python
import pandas as pd
import numpy as np

# Setting a random seed for reproducibility
np.random.seed(42)

# Number of instances, timepoints, and variables
n_instances = 3
n_timepoints = 5
n_variables = 4

# Generating random data
data = np.random.randn(n_instances * n_timepoints, n_variables)

# Create multiindex
arrays = [[f"instance_{i+1}" for i in range(n_instances) for _ in range(n_timepoints)],
          list(range(1, n_timepoints+1)) * n_instances]
index = pd.MultiIndex.from_arrays(arrays, names=('instances', 'timepoints'))

# Create DataFrame
df = pd.DataFrame(data, index=index, columns=[f"var_{i+1}" for i in range(n_variables)])

print(df)
```

```python
from sktime.transformations.panel.pca import PCATransformer
pca = PCATransformer()
pca.fit_transform(df)
```

```python
new_inner_indices = df.groupby(level=0).cumcount()
new_index = [(x[0], idx) for x, idx in zip(df.index.get_level_values(0), new_inner_indices)]
df.index = pd.MultiIndex.from_tuples(new_index, names=('instances', 'new_timepoints'))
df
```
